````python
# ===================== QUANTUM TUTOR AGENT CLASS =====================
print("\nğŸ¤– Defining Quantum Tutor Agent...")

class QuantumTutorAgent:
    """
    Adaptive AI Agent for Teaching Quantum Computing
    
    Teaches 5 core concepts:
    1. Bell States & Entanglement
    2. Grover's Search Algorithm
    3. Deutsch-Jozsa Algorithm
    4. Quantum Error Correction
    5. BB84 Quantum Cryptography
    """
    
    def __init__(self, api_key=None):
        """Initialize the Quantum Tutor Agent"""
        self.simulator = Aer.get_backend('qasm_simulator')
        self.statevector_sim = Aer.get_backend('statevector_simulator')
    
        # Google API Key setup
        self.api_key = api_key
        self.ai_enabled = False
    
        if api_key:
            try:
                import google.generativeai as genai
                genai.configure(api_key=api_key)
                self.model = genai.GenerativeModel('gemini-pro')
                self.ai_enabled = True
                print("âœ… AI Agent mode enabled with Google Gemini!")
            except Exception as e:
                print(f"âš ï¸  Could not initialize AI: {e}")
                print("   Running in standard tutorial mode")
        else:
            print("â„¹ï¸  Running in standard tutorial mode (no API key)")
    
        self.student_profile = {
            'topics_completed': [],
            'current_topic': None,
            'knowledge_level': 'beginner'
        }
    
    def get_ai_explanation(self, topic, question):
        """Get AI-powered personalized explanation"""
        if not self.ai_enabled:
            return None
    
        try:
            prompt = f"""You are a quantum computing tutor. The student is learning about {topic}.
            Student question: {question}
        
            Provide a clear, encouraging explanation in 2-3 paragraphs. 
            Use analogies and relate to real-world examples."""
        
            response = self.model.generate_content(prompt)
            return response.text
        except Exception as e:
            print(f"AI response error: {e}")
            return None
    # ==================== TEACHING FUNCTIONS ====================
    
    def teach_bell_states(self, detail_level="standard"):
        """Teach Bell states and entanglement"""
        print("\n" + "="*80)
        print(" "*25 + "ğŸ”— BELL STATES & ENTANGLEMENT")
        print("="*80 + "\n")
        
        # Create Bell state circuit
        qr = QuantumRegister(2, 'q')
        cr = ClassicalRegister(2, 'c')
        circuit = QuantumCircuit(qr, cr)
        
        circuit.h(qr[0])
        circuit.cx(qr[0], qr[1])
        circuit.measure(qr, cr)
        
        # Simulate
        job = self.simulator.run(transpile(circuit, self.simulator), shots=1000)
        counts = job.result().get_counts()
        
        # Draw circuit
        print("Quantum Circuit:")
        print("-"*80)
        print(circuit.draw(output='text', fold=-1))
        print("-"*80)
        
        explanation = f"""
ğŸ“š CONCEPT: Bell States - Quantum Entanglement

What is Entanglement?
Entanglement is a quantum phenomenon where two particles become correlated
in such a way that measuring one instantly affects the other, regardless
of the distance between them. Einstein called this "spooky action at a distance."

Circuit Construction:
1. Start with |00âŸ© (both qubits initialized to 0)
2. Apply Hadamard (H) to qubit 0
   Result: (|00âŸ© + |10âŸ©)/âˆš2 (qubit 0 in superposition)
3. Apply CNOT with control=q0, target=q1
   Result: (|00âŸ© + |11âŸ©)/âˆš2 â† This is a Bell state!

ğŸ“Š Simulation Results (1000 measurements):
{counts}

ğŸ” Key Observation:
Notice we ONLY get |00âŸ© or |11âŸ©, never |01âŸ© or |10âŸ©!
The qubits are perfectly correlated - they always match!

Why This Matters:
â€¢ Foundation for quantum teleportation
â€¢ Enables quantum cryptography
â€¢ Used in quantum error correction
â€¢ Proves quantum mechanics is complete (Bell's theorem)

ğŸ’» Qiskit Code:
```python
from qiskit import QuantumCircuit

# Create Bell state
qc = QuantumCircuit(2, 2)
qc.h(0)           # Hadamard on qubit 0
qc.cx(0, 1)       # CNOT: control=0, target=1
qc.measure_all()  # Measure both qubits

# Run on simulator
from qiskit_aer import Aer
simulator = Aer.get_backend('qasm_simulator')
job = simulator.run(qc, shots=1000)
result = job.result()
print(result.get_counts())
```

âœ¨ Fun Fact: 
Bell states were first discussed in the Einstein-Podolsky-Rosen (EPR) 
paradox paper in 1935. Einstein didn't believe in entanglement, but 
experiments in the 1980s proved it's real!
"""
    
        print(explanation)
        sys.stdout.flush()  # Force immediate display
        
        # Visualize results
        if counts:
            self._plot_histogram(counts, "Bell State Measurement Results")
        
        self.student_profile['topics_completed'].append('Bell States')
        print("\nâœ… Topic completed! Added to your learning profile.")
        return True
    
    def teach_grover(self, n_qubits=2):
        """Teach Grover's search algorithm"""
        print("\n" + "="*80)
        print(" "*25 + "ğŸ” GROVER'S SEARCH ALGORITHM")
        print("="*80 + "\n")
        
        target = '11'
        N = 2 ** n_qubits
        optimal_iters = int(np.pi / 4 * np.sqrt(N))
        
        # Create Grover circuit
        qr = QuantumRegister(n_qubits)
        cr = ClassicalRegister(n_qubits)
        circuit = QuantumCircuit(qr, cr)
        
        # Initialize superposition
        circuit.h(range(n_qubits))
        
        # Grover iterations
        for _ in range(optimal_iters):
            # Oracle
            circuit.cz(0, 1)
            
            # Diffusion operator
            circuit.h(range(n_qubits))
            circuit.x(range(n_qubits))
            circuit.h(n_qubits-1)
            if n_qubits == 2:
                circuit.cz(0, 1)
            circuit.h(n_qubits-1)
            circuit.x(range(n_qubits))
            circuit.h(range(n_qubits))
        
        circuit.measure_all()
        
        # Simulate
        job = self.simulator.run(transpile(circuit, self.simulator), shots=1000)
        counts = job.result().get_counts()
        
        print("Quantum Circuit:")
        print("-"*80)
        print(circuit.draw(output='text', fold=-1))
        print("-"*80)
        
        explanation = f"""
ğŸ“š CONCEPT: Grover's Algorithm - Quantum Search

The Problem:
Find a specific item in an UNSORTED database of N items.

Classical Approach:
Must check each item one by one â†’ O(N) time complexity
For {N} items: Need ~{N//2} queries on average

Quantum Approach:
Grover's algorithm â†’ O(âˆšN) time complexity
For {N} items: Need only {optimal_iters} iterations!

ğŸš€ Speedup: {N/optimal_iters:.1f}x faster!

How It Works:
1. INITIALIZATION: Put all qubits in equal superposition
   All {N} states have equal amplitude: 1/âˆš{N}

2. GROVER ITERATION (repeat {optimal_iters} times):
   a) ORACLE: Mark the target state by flipping its phase
   b) DIFFUSION: Amplify marked state, suppress others
      (This is called "amplitude amplification")

3. MEASUREMENT: Target state has highest probability!

ğŸ“Š Results (1000 measurements):
{counts}

Target state '{target}' found with ~{max(counts.values())/10:.0f}% probability!

Real-World Applications:
â€¢ Database search
â€¢ Optimization problems (traveling salesman, etc.)
â€¢ Breaking cryptography (searching for private keys)
â€¢ Machine learning (training optimization)
â€¢ SAT solvers (finding solutions to logical formulas)

ğŸ’» Qiskit Code:
```python
from qiskit import QuantumCircuit
import numpy as np

n = {n_qubits}
qc = QuantumCircuit(n, n)

# Initialize superposition
qc.h(range(n))

# Grover iteration ({optimal_iters} times)
for _ in range({optimal_iters}):
    # Oracle (mark target)
    qc.cz(0, 1)
    
    # Diffusion operator
    qc.h(range(n))
    qc.x(range(n))
    qc.cz(0, 1)
    qc.x(range(n))
    qc.h(range(n))

qc.measure_all()
```

âœ¨ Fun Fact:
Lov Grover invented this algorithm in 1996. It's one of the most
practical quantum algorithms and will be used in real quantum computers!
"""
    
        print(explanation)
        sys.stdout.flush()  # Force immediate display
        
        if counts:
            self._plot_histogram(counts, "Grover's Algorithm Results")
        
        self.student_profile['topics_completed'].append('Grovers Algorithm')
        print("\nâœ… Topic completed! Added to your learning profile.")
        return True
    
    def teach_deutsch_jozsa(self, function_type="balanced"):
        """Teach Deutsch-Jozsa algorithm"""
        print("\n" + "="*80)
        print(" "*25 + "ğŸ¯ DEUTSCH-JOZSA ALGORITHM")
        print("="*80 + "\n")
        
        n = 2
        qr = QuantumRegister(n + 1)
        cr = ClassicalRegister(n)
        circuit = QuantumCircuit(qr, cr)
        
        # Initialize
        circuit.x(n)
        circuit.h(range(n + 1))
        
        # Oracle
        if function_type == "constant":
            pass  # Do nothing for constant-0
        else:
            # Balanced function
            circuit.cx(0, n)
            circuit.cx(1, n)
        
        circuit.h(range(n))
        circuit.measure(range(n), cr)
        
        # Simulate
        job = self.simulator.run(transpile(circuit, self.simulator), shots=1)
        result = list(job.result().get_counts().keys())[0]
        is_constant = result == '0' * n
        
        print("Quantum Circuit:")
        print("-"*80)
        print(circuit.draw(output='text', fold=-1))
        print("-"*80)
        
        explanation = f"""
ğŸ“š CONCEPT: Deutsch-Jozsa Algorithm - Quantum Parallelism

The Problem:
Given a function f: {{0,1}}â¿ â†’ {{0,1}}, determine if it is:
â€¢ CONSTANT: f(x) = 0 for all x, OR f(x) = 1 for all x
â€¢ BALANCED: f(x) = 0 for half inputs, f(x) = 1 for other half

Classical Approach:
Worst case: Must check 2^(n-1) + 1 inputs
For {n} qubits: Need up to {2**(n-1) + 1} function evaluations

Quantum Approach:
Deutsch-Jozsa algorithm: Only 1 query!
ğŸš€ Speedup: EXPONENTIAL!

How It Works:
1. Prepare ALL inputs in superposition
2. Query oracle ONCE (evaluates all inputs in parallel!)
3. Apply Hadamard transform
4. Measure:
   â€¢ All zeros â†’ Function is CONSTANT
   â€¢ Anything else â†’ Function is BALANCED

Oracle Type: {function_type.upper()}
Measurement Result: |{result}âŸ©
Algorithm Conclusion: {'CONSTANT' if is_constant else 'BALANCED'} âœ…

ğŸ“Š Why It Works:
The algorithm uses QUANTUM INTERFERENCE:
â€¢ If constant: All computational paths interfere constructively
  â†’ Measure |00...0âŸ©
â€¢ If balanced: Destructive interference on |00...0âŸ©
  â†’ Measure something else

Key Insight - Quantum Parallelism:
The superposition state represents ALL possible inputs simultaneously!
The oracle evaluates the function on all 2â¿ inputs in one query.
This is the power of quantum computing!

ğŸ’» Qiskit Code:
```python
from qiskit import QuantumCircuit

n = 2  # Input qubits
qc = QuantumCircuit(n+1, n)

# Prepare
qc.x(n)           # Ancilla to |1âŸ©
qc.h(range(n+1))  # All in superposition

# Oracle (balanced example)
qc.cx(0, n)
qc.cx(1, n)

# Measure
qc.h(range(n))
qc.measure(range(n), range(n))
```

âœ¨ Fun Fact:
This was one of the first algorithms to demonstrate quantum advantage!
David Deutsch proposed it in 1985, and Richard Jozsa generalized it in 1992.
"""

        print(explanation)
        sys.stdout.flush()  # Force immediate display
        
        self.student_profile['topics_completed'].append('Deutsch-Jozsa')
        print("\nâœ… Topic completed! Added to your learning profile.")
        return True
    
    def teach_error_correction(self):
        """Teach quantum error correction"""
        print("\n" + "="*80)
        print(" "*22 + "ğŸ›¡ï¸ QUANTUM ERROR CORRECTION")
        print("="*80 + "\n")
        
        explanation = """
ğŸ“š CONCEPT: Quantum Error Correction - Protecting Quantum Information

The Challenge:
Quantum states are EXTREMELY fragile! They suffer from:
â€¢ Decoherence (interaction with environment)
â€¢ Gate errors (imperfect quantum operations)
â€¢ Measurement errors
â€¢ Noise and interference

Types of Quantum Errors:
1. BIT FLIP: |0âŸ© â†” |1âŸ© (like classical bit flip)
2. PHASE FLIP: |+âŸ© â†” |-âŸ© (purely quantum!)
3. BOTH (arbitrary single-qubit errors)

The Paradox:
â€¢ We need to detect errors
â€¢ But we can't measure quantum states (destroys superposition!)
â€¢ We can't copy quantum states (no-cloning theorem)

The Solution: Quantum Error Correction Codes!

â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”
3-QUBIT BIT FLIP CODE (Simplest Example)
â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”

Encoding (1 logical qubit â†’ 3 physical qubits):
|0âŸ©_L â†’ |000âŸ©
|1âŸ©_L â†’ |111âŸ©
General state: Î±|0âŸ© + Î²|1âŸ© â†’ Î±|000âŸ© + Î²|111âŸ©

Error Detection (Syndrome Measurement):
Measure parity of qubit pairs WITHOUT measuring qubits themselves!
â€¢ Syndrome Zâ‚€Zâ‚: Are qubits 0 and 1 the same?
â€¢ Syndrome Zâ‚Zâ‚‚: Are qubits 1 and 2 the same?

Syndrome Table:
â”Œâ”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”¬â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”¬â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”
â”‚ Syndrome â”‚   Error Type    â”‚  Correction  â”‚
â”œâ”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”¼â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”¼â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”¤
â”‚   00     â”‚   No error      â”‚  Do nothing  â”‚
â”‚   01     â”‚ Error on qubit 2â”‚  Flip qubit 2â”‚
â”‚   10     â”‚ Error on qubit 0â”‚  Flip qubit 0â”‚
â”‚   11     â”‚ Error on qubit 1â”‚  Flip qubit 1â”‚
â””â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”´â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”´â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”˜

Circuit Diagram:
     â”Œâ”€â”€â”€â”     â”Œâ”€â”€â”€â”€â”€â”€â”€â”€â”€â”     â”Œâ”€â”€â”€â”
q_0: â”¤ H â”œâ”€â”€â– â”€â”€â”¤ X (err) â”œâ”€â”€â”€â”€â”€â”¤ X â”œâ”€â”€ (corrected)
     â””â”€â”€â”€â”˜â”Œâ”€â”´â”€â”â””â”€â”€â”€â”€â”€â”€â”€â”€â”€â”˜     â””â”€â”€â”€â”˜
q_1: â”€â”€â”€â”€â”€â”¤ X â”œâ”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€ 
          â””â”€â”€â”€â”˜   â”Œâ”€â”€â”€â”         â”Œâ”€â”€â”€â”
q_2: â”€â”€â”€â”€â”€â”€â”€â”€â– â”€â”€â”€â”€â”¤ M â”œâ”€â”€â”€â”€â”€â”€â”€â”€â”€â”¤ M â”œ (syndrome)
             â”‚    â””â”€â”€â”€â”˜         â””â”€â”€â”€â”˜
             â””â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€

Key Insight:
We detect and correct errors WITHOUT collapsing the quantum state!
The syndrome measurement is designed to be "gentle" - it tells us
about the ERROR without revealing the QUBIT VALUE.

Advanced Error Correction Codes:
â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”
1. SHOR CODE (9 qubits)
   â€¢ Corrects any single qubit error (bit flip AND phase flip)
   â€¢ First quantum error correction code (1995)

2. STEANE CODE (7 qubits)
   â€¢ CSS (Calderbank-Shor-Steane) code
   â€¢ More efficient than Shor code

3. SURFACE CODES (many qubits)
   â€¢ Used in real quantum computers (Google, IBM)
   â€¢ Scalable to large quantum computers
   â€¢ Error rate threshold: ~1%

4. TOPOLOGICAL CODES
   â€¢ Use 2D lattice of qubits
   â€¢ Errors are "anyons" that can be tracked

ğŸ’» Qiskit Code (3-Qubit Bit Flip):
```python
from qiskit import QuantumCircuit

qc = QuantumCircuit(5, 2)  # 3 data + 2 ancilla

# Encode logical |+âŸ©
qc.h(0)
qc.cx(0, 1)
qc.cx(0, 2)

# Error occurs (simulate)
qc.x(1)  # Bit flip on qubit 1

# Syndrome measurement
qc.cx(0, 3)  # Ancilla 0
qc.cx(1, 3)
qc.cx(1, 4)  # Ancilla 1
qc.cx(2, 4)
qc.measure([3, 4], [0, 1])

# Correction based on syndrome
# (requires classical feedback)

# Decode
qc.cx(0, 2)
qc.cx(0, 1)
qc.h(0)
```

The Future:
â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”
â€¢ Current quantum computers have error rates of 0.1% - 1%
â€¢ Need error rates < 0.01% for useful computations
â€¢ Fault-tolerant quantum computing requires ~1000 physical qubits
  per logical qubit with surface codes
â€¢ Goal: Million-qubit quantum computers with error correction

âœ¨ Fun Fact:
Peter Shor's 9-qubit code (1995) was revolutionary because it showed
quantum error correction is POSSIBLE! Before this, many thought quantum
computing was impossible due to decoherence.
"""

        print(explanation)
        sys.stdout.flush()  # Force immediate display
        
        self.student_profile['topics_completed'].append('Error Correction')
        print("\nâœ… Topic completed! Added to your learning profile.")
        return True
    
    def teach_bb84(self, key_length=8):
        """Teach BB84 quantum cryptography"""
        print("\n" + "="*80)
        print(" "*22 + "ğŸ” BB84 QUANTUM CRYPTOGRAPHY")
        print("="*80 + "\n")
        
        # Simulate BB84 protocol
        alice_bits = np.random.randint(0, 2, key_length)
        alice_bases = np.random.randint(0, 2, key_length)
        bob_bases = np.random.randint(0, 2, key_length)
        
        bob_results = []
        matching = []
        
        for i in range(key_length):
            if alice_bases[i] == bob_bases[i]:
                bob_results.append(alice_bits[i])
                matching.append(True)
            else:
                bob_results.append(np.random.randint(0, 2))
                matching.append(False)
        
        sifted_key = [alice_bits[i] for i in range(key_length) if matching[i]]
        
        print("BB84 Protocol Simulation:")
        print("-"*80)
        print(f"Key Length: {key_length} qubits")
        print(f"\nAlice's bits:   {alice_bits.tolist()}")
        print(f"Alice's bases:  {['Z' if b==0 else 'X' for b in alice_bases]}")
        print(f"Bob's bases:    {['Z' if b==0 else 'X' for b in bob_bases]}")
        print(f"Bases match?    {['âœ“' if m else 'âœ—' for m in matching]}")
        print(f"\nSifted key:     {sifted_key}")
        print(f"Final key length: {len(sifted_key)} bits")
        print(f"Efficiency:     {100*len(sifted_key)/key_length:.0f}%")
        print("-"*80)
        
        explanation = f"""
ğŸ“š CONCEPT: BB84 - Quantum Key Distribution

The Goal:
Securely share a secret cryptographic key between Alice and Bob
over an INSECURE channel, with provable security!

Why BB84 is Revolutionary:
â€¢ Information-theoretically secure (not just computationally)
â€¢ Security based on laws of physics, not math assumptions
â€¢ Eavesdropping is detectable (thanks to quantum mechanics!)
â€¢ Secure even against quantum computers!

â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”
THE BB84 PROTOCOL (Bennett & Brassard, 1984)
â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”

Step 1: ALICE PREPARES QUBITS
For each bit she wants to send:
â€¢ Randomly choose bit value: 0 or 1
â€¢ Randomly choose basis:
  - Rectilinear (Z): |0âŸ© or |1âŸ©
  - Diagonal (X): |+âŸ© = (|0âŸ©+|1âŸ©)/âˆš2 or |-âŸ© = (|0âŸ©-|1âŸ©)/âˆš2

Encoding Table:
â”Œâ”€â”€â”€â”€â”€â”€â”¬â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”¬â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”
â”‚ Bit  â”‚  Basis   â”‚    State     â”‚
â”œâ”€â”€â”€â”€â”€â”€â”¼â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”¼â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”¤
â”‚  0   â”‚    Z     â”‚     |0âŸ©      â”‚
â”‚  1   â”‚    Z     â”‚     |1âŸ©      â”‚
â”‚  0   â”‚    X     â”‚ |+âŸ© (H|0âŸ©)   â”‚
â”‚  1   â”‚    X     â”‚ |-âŸ© (H|1âŸ©)   â”‚
â””â”€â”€â”€â”€â”€â”€â”´â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”´â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”˜

Step 2: ALICE SENDS QUBITS
Qubits travel through quantum channel (could be fiber optic cable)

Step 3: BOB MEASURES
For each qubit:
â€¢ Randomly choose measurement basis (Z or X)
â€¢ Measure in chosen basis
â€¢ Record result

Step 4: BASIS RECONCILIATION (Public Channel)
â€¢ Alice and Bob publicly announce their bases (NOT the bits!)
â€¢ Keep bits where bases matched, discard others
â€¢ This creates the "sifted key" ({len(sifted_key)} bits in our example)

Step 5: ERROR CHECKING
â€¢ Compare a random subset of sifted key
â€¢ If error rate > threshold â†’ eavesdropper detected!
â€¢ If error rate low â†’ proceed

Step 6: PRIVACY AMPLIFICATION
â€¢ Use remaining bits to create final secret key

â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”
WHY IT'S SECURE: QUANTUM PHYSICS GUARANTEES
â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”

1. NO-CLONING THEOREM
   Eve cannot copy quantum states perfectly

2. MEASUREMENT DISTURBS STATES
   If Eve measures a qubit in wrong basis:
   â€¢ She gets random result
   â€¢ The qubit state changes
   â€¢ Alice & Bob detect increased error rate

3. HEISENBERG UNCERTAINTY
   Eve cannot know both Z and X measurements simultaneously

Example Attack:
Eve intercepts qubit in state |+âŸ©
â€¢ If Eve measures in Z basis: Gets 0 or 1 (random!)
â€¢ Qubit collapses to |0âŸ© or |1âŸ©
â€¢ Bob now measures different state than Alice sent
â€¢ Error detected!

Real-World Implementations:
â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”â”
â€¢ ID Quantique (Switzerland): Commercial QKD systems since 2001
â€¢ China's Micius Satellite: Quantum communication over 1200 km (2017)
â€¢ DARPA: Quantum network in Boston (2003-2007)
â€¢ Banks: Used for secure transaction processing
â€¢ Governments: Secure diplomatic communications

Limitations:
â€¢ Distance limited to ~100-200 km (fiber optics)
â€¢ Requires quantum repeaters for long distances
â€¢ Slow key generation rate
â€¢ Expensive equipment

The Future:
â€¢ Quantum internet connecting quantum computers
â€¢ Satellite-based global quantum network
â€¢ Integration with 5G/6G networks

ğŸ’» Qiskit Code:
```python
from qiskit import QuantumCircuit

def alice_prepare(bit, basis):
    qc = QuantumCircuit(1, 1)
    if bit == 1:
        qc.x(0)  # Encode bit
    if basis == 1:  # X basis
        qc.h(0)  # Apply Hadamard
    return qc

def bob_measure(qc, basis):
    if basis == 1:  # X basis
        qc.h(0)  # Apply Hadamard before measuring
    qc.measure(0, 0)
    return qc

# Alice prepares |+âŸ© (bit=0, basis=X)
qc = alice_prepare(0, 1)
# Bob measures in X basis (match!)
qc = bob_measure(qc, 1)
# Result: 0 with high probability
```

âœ¨ Fun Fact:
BB84 was proposed in 1984 but largely ignored until the 1990s.
Now it's used in real quantum networks! Charles Bennett and
Gilles Brassard's protocol was ahead of its time.
"""

        print(explanation)
        sys.stdout.flush()  # Force immediate display
        
        self.student_profile['topics_completed'].append('BB84')
        print("\nâœ… Topic completed! Added to your learning profile.")
        return True
    
    def show_menu(self):
        """Display interactive menu"""
        print("\n" + "="*80)
        print(" "*20 + "ğŸŒŸ QUANTUM COMPUTING TUTOR MENU ğŸŒŸ")
        print("="*80)
        print("\nğŸ“š Available Topics:\n")
        print("  [1] ğŸ”— Bell States & Quantum Entanglement (Beginner)")
        print("      Learn about superposition and quantum correlations")
        print()
        print("  [2] ğŸ” Grover's Search Algorithm (Intermediate)")
        print("      Discover quantum speedup for database search")
        print()
        print("  [3] ğŸ¯ Deutsch-Jozsa Algorithm (Intermediate)")
        print("      See quantum parallelism in action")
        print()
        print("  [4] ğŸ›¡ï¸ Quantum Error Correction (Advanced)")
        print("      Learn to protect quantum information")
        print()
        print("  [5] ğŸ” BB84 Quantum Cryptography (Advanced)")
        print("      Explore provably secure communication")
        print()
        print("  [6] ğŸ“Š Show My Progress")
        print("      View completed topics and statistics")
        print()
        print("  [0] ğŸ‘‹ Exit Tutor")
        print("="*80)
    
    def show_progress(self):
        """Show student progress"""
        print("\n" + "="*80)
        print(" "*25 + "ğŸ“Š YOUR LEARNING PROGRESS")
        print("="*80)
        
        completed = self.student_profile['topics_completed']
        total = 5
        
        print(f"\nâœ… Topics Completed: {len(completed)}/{total}")
        print(f"ğŸ“ˆ Progress: {len(completed)/total*100:.0f}%")
        print("\n" + "-"*80)
        
        all_topics = [
            'Bell States',
            'Grovers Algorithm', 
            'Deutsch-Jozsa',
            'Error Correction',
            'BB84'
        ]
        
        for topic in all_topics:
            status = "âœ…" if topic in completed else "â¬œ"
            print(f"{status} {topic}")
        
        print("-"*80)
        
        if len(completed) == total:
            print("\nğŸ‰ CONGRATULATIONS! You've completed all quantum computing topics!")
            print("   You're now ready to build quantum algorithms!")
        elif len(completed) >= 3:
            print("\nğŸŒŸ Great progress! You're well on your way to mastering quantum computing!")
        elif len(completed) >= 1:
            print("\nğŸ‘ Good start! Keep learning to unlock more quantum concepts!")
        else:
            print("\nğŸ’¡ Start your quantum journey! Pick a topic from the menu.")
    print("\n" + "="*80)

    def _plot_histogram(self, counts, title):
        """Plot measurement results"""
        try:
            plt.figure(figsize=(10, 6))
            states = list(counts.keys())
            values = list(counts.values())
        
            colors = ['#1f77b4' if i % 2 == 0 else '#ff7f0e' for i in range(len(states))]
            plt.bar(states, values, color=colors, alpha=0.8, edgecolor='black')
        
            plt.xlabel('Quantum State', fontsize=12, fontweight='bold')
            plt.ylabel('Count (out of 1000)', fontsize=12, fontweight='bold')
            plt.title(title, fontsize=14, fontweight='bold')
            plt.grid(axis='y', alpha=0.3)
        
            # Add value labels on bars
            for i, (state, value) in enumerate(zip(states, values)):
                plt.text(i, value + 10, str(value), ha='center', fontweight='bold')
        
            plt.tight_layout()
            plt.show()
        except Exception as e:
            print(f"Note: Could not generate plot - {e}")
    print("âœ… Quantum Tutor Agent class defined successfully!")
````
